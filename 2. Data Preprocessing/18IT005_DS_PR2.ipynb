```python
#Assigning column names for the dataset
headernames = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']
#importing dataset
dataset = pd.read_csv('iris.data', names = headernames)
dataset.head()
```

```python
#Class before encoding
print("Types of classes and their count:\n",dataset['Class'].value_counts())
```

# **Encoding**

```python
# Encoding Using Label Encoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset['Class']=le.fit_transform(dataset['Class'])

print("Classes: ",le.classes_)
dataset
```

```python
# Encoding Using OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(dataset['Class'].values.reshape(-1,1)).toarray()
print("Encoded Categories: ",encoder.categories_)

#Transforming the encoded data to dataframe
encoded_data = pd.DataFrame(encoded_data , columns = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
encoded_data
```

**Standardization**

```python
dataset
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
copydata = dataset
#Performing scaling using StandardScaler
sc = StandardScaler()
standardized_data = sc.fit_transform(copydata)
print("Before Standardization\n")
print(dataset)
print("After Standardization\n")
pd.DataFrame(standardized_data,columns=copydata.columns)
```

Normalization

```python
#Normalization
from sklearn.preprocessing import normalize
copydata1 = dataset.drop(['Class'], axis = 'columns')
normalized = normalize(copydata1, norm='l2')
print("Before Normalization\n")
print(dataset)
print("After Normalization\n")
pd.DataFrame(normalized,columns=copydata1.columns)
```

Discretization

```python
from sklearn.preprocessing import KBinsDiscretizer
kbins = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='uniform')
copydata4 = dataset.drop(['Class'], axis = 'columns')
data_trans = kbins.fit_transform(copydata4)

pd.DataFrame(copydata4,columns=copydata4.columns)
```
